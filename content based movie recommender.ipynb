```python
def str_to_dict(str):
    lst=ast.literal_eval(str)
    for i in lst:
        if(i['job']=='Director'):
            return (i['name'])
```

```python
def extract_genres(str):
    lst=ast.literal_eval(str)
    anslst=[]
    for i in lst:
        anslst.append(i["name"])
    return anslst
```

```python
extract_genres(movies_df.genres.iloc[1])
```

```python
movies_df['genres']=movies_df['genres'].apply(extract_genres)
# movies_df['genres']=pd.read
```

```python
movies_df['keywords']=movies_df['keywords'].apply(extract_genres)
```

```python
def extract_cast(str):
    lst=ast.literal_eval(str)
    anslst=[]
    count=0
    for i in lst:
        count+=1
        if(count>3):
            break
        anslst.append(i["name"])
    return anslst
```

```python
movies_df['cast']=(movies_df.cast).apply(extract_cast)
```

```python

def space_remove(lst):
    anslst=[]
    for j in lst:
        anslst.append(j.replace(' ',''))
    return anslst
```

```python
movies_df['genres']=movies_df['genres'].apply(space_remove)
```

```python
movies_df['keywords']=movies_df['keywords'].apply(space_remove)
movies_df['cast']=movies_df['cast'].apply(space_remove)
```

```python
def crew_handle(lst):
    ansstr=""
    anslst=[];
    for i in lst:
        # anslst.append(i)
        ansstr+=i
    anslst.append(ansstr)
    return anslst
```

```python
movies_df['crew']=movies_df['crew'].apply(crew_handle)
```

```python
df=movies_df.copy()
movies_df['tags']=movies_df['genres']+movies_df['keywords']+movies_df['overview']+movies_df['cast']+movies_df['crew']
movies_df.sample()
```

```python
movies_df=movies_df.drop(axis=1,labels=['genres','keywords','overview','cast','crew'])
```

```python
import nltk
ps = nltk.stem.PorterStemmer()
```

```python
df=movies_df.copy()
movies_df.tags=movies_df.tags.apply(ps.stem)
movies_df.tags.iloc[0]
```

```python
df=movies_df.copy()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
list(enumerate(similarity[0]))
```

```python
index=df[df['title']=='John Carter'].index[0]
index
```

```python
def recommender(movie):
    index=df[df['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])[1:6]
    for i in distances:
        print(df.iloc[i[0]].title)
```

```python
recommender('John Carter')
# vector similarity
# distance
# alss 
# eval metrics
# 
```

```python
recommender('Aliens vs Predator: Requiem')
```
